# Creating a decision tree for Task 2 

## 1. Sample code - Iris dataset 

### Importing...

```python
from sklearn.datasets import load_iris
import sklearn
from sklearn import tree
import pandas as pd
import os
import graphviz 
```

## Creating Decision tree

```python
iris = load_iris()
clf = tree.DecisionTreeClassifier()
clf1 = clf.fit(iris.data, iris.target)

clf1.predict(iris.data)
clf1.tree_.__getstate__()
```

```python
iris.target
```

## Graphing

```python
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
graph.render("iris")
```
